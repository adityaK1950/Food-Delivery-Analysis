```python
import pandas as pd

df = pd.read_csv("food delivery costs.csv")
df



```

```python
df.head()
```

```python
df.info()
```

```python
# Cleaning
df["Order Date and Time"] = pd.to_datetime(df['Order Date and Time'])
df.info()
```

```python
df["Delivery Date and Time"] = pd.to_datetime(df['Delivery Date and Time'])
df.info()
```

```python
df.head()
```

```python
def extract(value):
    a = str(value).split(" ")
    return a[0]

df["Discounts and Offers"] = df["Discounts and Offers"].apply(extract)
df.head()
```

```python
def removep(value):
    if "%" in value:
        a = value.replace("%","")
        return float(a)
    else:
        return float(value)

df["Discounts and Offers"] = df["Discounts and Offers"].apply(removep)
df.head()
```

```python
df.loc[(df["Discounts and Offers"] <= 15),"Discounts and Offers" ] = (df["Discounts and Offers"]/100) * df["Order Value"]
df.head()
```

```python
df["Discounts and Offers"] = df["Discounts and Offers"].fillna(0)
df.head()
```

```python
df["Costs"] = df["Delivery Fee"] + df['Discounts and Offers'] + df["Payment Processing Fee"]
df.head()
```

```python
df["Profit"] = df["Commission Fee"] - df['Costs']
df.head()
```

```python
df["Profit"].sum()
```

```python
cost_dist = df[["Delivery Fee", "Payment Processing Fee", "Discounts and Offers"]].sum()
cost_dist
```

```python
profit_by_payment_method = df.groupby("Payment Method")["Profit"].mean().reset_index()
print(profit_by_payment_method)
```

```python
df['Order Date'] = df['Order Date and Time'].dt.date
df['Order Hour'] = df['Order Date and Time'].dt.hour
profit_by_hour = df.groupby("Order Hour")["Profit"].mean().reset_index()
print(profit_by_hour)
```

```python
customer_segment = df.groupby("Customer ID").agg({"Order Value": "sum", "Profit": "sum", "Order ID": "count"})
customer_segment = customer_segment.rename(columns={"Order ID": "Order Frequency"})
print(customer_segment.sort_values(by="Order Value", ascending=False).head())
```

```python
discount_effectiveness = df.groupby("Discounts and Offers")["Profit"].mean().reset_index()
print(discount_effectiveness)
```

```python
refund_analysis = df.groupby("Refunds/Chargebacks")["Profit"].mean().reset_index()
print(refund_analysis)
```

```python
df["Delivery Time (Minutes)"] = (df["Delivery Date and Time"] - df["Order Date and Time"]).dt.total_seconds() / 60
delivery_time_analysis = df.groupby("Delivery Time (Minutes)")["Profit"].mean().reset_index()
print(delivery_time_analysis)
```

```python
# Convert date columns to datetime format
df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time'])
df['Delivery Date and Time'] = pd.to_datetime(df['Delivery Date and Time'])

```

```python
# Clean 'Discounts and Offers'
def extract(value):
    if pd.isna(value):
        return 0
    a = str(value).split(" ")
    return a[0]

df["Discounts and Offers"] = df["Discounts and Offers"].apply(extract)
```

```python

def removep(value):
    if "%" in value:
        a = value.replace("%","")
        return float(a)
    else:
        return float(value)

df["Discounts and Offers"] = df["Discounts and Offers"].apply(removep)
df["Discounts and Offers"] = df["Discounts and Offers"].fillna(0)
```

```python

# Calculate Costs and Profit
df["Costs"] = df["Delivery Fee"] + df['Discounts and Offers'] + df["Payment Processing Fee"]
df["Profit"] = df["Commission Fee"] - df['Costs']
```

```python
# 1. Distribution of Order Values
plt.figure(figsize=(10, 6))
sns.histplot(df['Order Value'], kde=True, bins=30)
plt.title('Distribution of Order Values')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Distribution of Profits
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit'], kde=True, bins=30, color='red')
plt.title('Distribution of Profits')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Order Value vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Order Value', y='Profit', data=df, alpha=0.5)
plt.title('Order Value vs Profit')
plt.xlabel('Order Value')
plt.ylabel('Profit')
plt.show()
```

```python
# 4. Total Costs Distribution
total_costs = df[['Delivery Fee', 'Payment Processing Fee', 'Discounts and Offers']].sum()
plt.figure(figsize=(6, 6))
plt.pie(total_costs, labels=total_costs.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Total Costs')
plt.show()
```

```python
# 5. Commission Fee vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Commission Fee', y='Profit', data=df, alpha=0.5)
plt.title('Commission Fee vs Profit')
plt.xlabel('Commission Fee')
plt.ylabel('Profit')
plt.show()
```

```python
# 6. Profits by Payment Method
plt.figure(figsize=(12, 8))
sns.boxplot(x='Payment Method', y='Profit', data=df)
plt.title('Profit by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Monthly Order Value
df['Month'] = df['Order Date and Time'].dt.to_period('M')
monthly_order_value = df.groupby('Month')['Order Value'].sum()
plt.figure(figsize=(12, 6))
monthly_order_value.plot(kind='bar')
plt.title('Total Order Value by Month')
plt.xlabel('Month')
plt.ylabel('Total Order Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Average Delivery Fee by Restaurant

top_n = 20  # Change to 5 if you want the top 5 restaurants

avg_delivery_fee_by_restaurant = df.groupby('Restaurant ID')['Delivery Fee'].mean().sort_values(ascending=False).head(top_n)

plt.figure(figsize=(12, 8))
avg_delivery_fee_by_restaurant.plot(kind='barh', color='skyblue')
plt.title(f'Top {top_n} Restaurants by Average Delivery Fee')
plt.xlabel('Average Delivery Fee')
plt.ylabel('Restaurant ID')
plt.show()
```

```python
# Group by Restaurant ID and sum the Order Value
order_value_by_restaurant = df.groupby('Restaurant ID')['Order Value'].sum().reset_index()
```

```python
# Sort the result (optional)
order_value_by_restaurant = order_value_by_restaurant.sort_values(by='Order Value', ascending=False)
```

```python
# Visualize the result

plt.figure(figsize=(10, 6))
sns.barplot(x='Order Value', y='Restaurant ID', data=order_value_by_restaurant.head(10), hue='Restaurant ID', palette='Set2', legend=False)
plt.title('Top 10 Order Values by Restaurant')
plt.xlabel('Total Order Value')
plt.ylabel('Restaurant ID')
plt.show()
```

```python
# 9. Profit by Restaurant (Top 10)
top_n = 15  # Change to 5 if you want the top 5 restaurants

# Calculate average profit by restaurant and get the top N restaurants
top_restaurants = df.groupby('Restaurant ID')['Profit'].mean().sort_values(ascending=False).head(top_n).index

# Filter the original DataFrame to include only the top N restaurants
filtered_df = df[df['Restaurant ID'].isin(top_restaurants)]

plt.figure(figsize=(12, 8))
sns.boxplot(x='Restaurant ID', y='Profit', data=filtered_df, hue='Restaurant ID', palette='Set2', legend=False)
plt.title(f'Profit by Top {top_n} Restaurants')
plt.xlabel('Restaurant ID')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```
